```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
fx = []
x1 = []
x2 = []

with open('steps.txt', 'r') as f:
    for line in f:
        items = line.split('\t')
        fx.append(items[0].split(' ')[1])
        x1.append(items[1].split(' ')[1])
        x2.append(items[2].split(' ')[1])
        
        
print(fx)
print(x1)
print(x2)
```

```python
print(len(fx))
```

```python
fx = np.array(fx)
x1 = np.array(x1)
x2 = np.array(x2)
```

```python

fx = [float(i) for i in fx]
    
#s = 1 + np.sin(2 * np.pi * t)
s = range(len(fx))
# create a figure and axes
fig, ax = plt.subplots(figsize=(30, 15))
  
ax.set_title('Example Graph')
  
ax.set_ylabel('f(x)')
ax.set_xlabel('X1')
  
# set x, y-axis limits 
ax.set_xlim(min(fx), max(fx))
ax.set_ylim(min(s), max(s))
  
# function to plot and show graph
ax.plot(s, fx)
plt.show()
```
